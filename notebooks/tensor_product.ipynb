```python
# Bring in Qiskit library so we can use quantum tools.
import qiskit

# Import Statevector to represent quantum states, and Operator to represent gates.
from qiskit.quantum_info import Statevector, Operator
# Import square-root function for normalized quantum numbers.
from numpy import sqrt

# Print which Qiskit version is running in this notebook.
print("Qiskit version:", qiskit.__version__)

# Create the basic |0> state (like "off" for one qubit).
zero = Statevector.from_label('0')
# Create the basic |1> state (like "on" for one qubit).
one = Statevector.from_label('1')

# Show |0> in pretty math format.
display(zero.draw("latex"))
# Show |1> in pretty math format.
display(one.draw("latex"))

# Combine |0> and |1> into a 2-qubit state using tensor product.
psi = zero.tensor(one)
# Show the combined 2-qubit state.
display(psi.draw("latex"))

# Create |+> state (equal mix of |0> and |1>).
plus = Statevector.from_label('+')
# Create |-> state (equal mix with minus sign).
minus = Statevector.from_label('-')

# Show |+> state.
display(plus.draw("latex"))
# Show |-> state.
display(minus.draw("latex"))

# Create right-circular basis state (same idea people often call +i).
plusi = Statevector.from_label('r')
# Create left-circular basis state (same idea people often call -i).
minusi = Statevector.from_label('l')

# Show right-circular state.
display(plusi.draw("latex"))
# Show left-circular state.
display(minusi.draw("latex"))

# Combine |+> with left-circular state into another 2-qubit state.
phi = plus.tensor(minusi)
# Show the new combined state.
display(phi.draw("latex"))

# Build Hadamard gate operator.
H = Operator.from_label('H')
# Build identity gate (does nothing).
Id = Operator.from_label('I')
# Build Pauli-X gate (bit-flip gate).
X = Operator.from_label('X')
# Combine gates for two qubits: H on first qubit, I on second qubit.
display(H.tensor(Id).draw("latex"))

# Apply X gate to |+> (state evolves under the operator), then show result.
display(plus.evolve(X).draw("latex"))


w = Statevector([0, 1, 1, 0, 1, 0, 0, 0] / sqrt(3))
display(w.draw("latex"))

result, state = w.measure([0])
print(f"Measured: {result}\nState after measurement:")
display(state.draw("latex"))

result, state = w.measure([0, 1])
print(f"Measured: {result}\nState after measurement:")
display(state.draw("latex"))
```

```python
valid_labels = ['0', '1', '+', '-', 'r', 'l']

print("Valid Statevector labels in this Qiskit version:", valid_labels)

for label in valid_labels:
    state = Statevector.from_label(label)
    print(f"\nLabel '{label}' as text:")
    display(state.draw("text"))
    print(f"Label '{label}' as LaTeX:")
    display(state.draw("latex"))
```
